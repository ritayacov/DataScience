### Cramer's V - a correlation measure for categorial variables based on ChiSquare

##### Reference: <br>
Wikipedia - https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V <br>
R-Bloggers - https://www.r-bloggers.com/2011/06/example-8-39-calculating-cramers-v/

```r
cv.test = function(x,y) {
  dd = chisq.test(x, y, correct=FALSE)
  CV = sqrt(dd$statistic /
    (length(x) * (min(length(unique(x)),length(unique(y))) - 1)))
  #print.noquote("Cramér V / Phi:")
  return(cbind('Cramér V'=as.numeric(CV),'p-value'=dd$p.value))
}
```

```r
cats_corr <- NULL
cats_corr <- rbind(cats_corr, cbind("WindGustDir", "WindDir9am",(cv.test(df$WindGustDir, df$WindDir9am))))
cats_corr <- rbind(cats_corr, cbind("WindGustDir", "WindDir3am",(cv.test(df$WindGustDir, df$WindDir3pm))))
cats_corr <- rbind(cats_corr, cbind("WindDir3pm", "WindDir9am",(cv.test(df$WindDir3pm, df$WindDir9am))))
cats_corr <- rbind(cats_corr, cbind("RainToday", "RainTomorrow",(cv.test(df$RainToday, df$RainTomorrow))))
cats_corr <- rbind(cats_corr, cbind("WindGustDir", "RainTomorrow",(cv.test(df$WindGustDir, df$RainTomorrow))))
cats_corr <- rbind(cats_corr, cbind("WindDir3pm", "RainTomorrow",(cv.test(df$WindDir3pm, df$RainTomorrow))))
cats_corr <- rbind(cats_corr, cbind("WindDir9pm", "RainTomorrow",(cv.test(df$WindDir9am, df$RainTomorrow))))
cats_corr <- data.frame(cats_corr)
row.names(cats_corr) <- NULL
cats_corr$Cramér.V <- round(as.numeric(cats_corr$Cramér.V),5)
cats_corr$p.value <- round(as.numeric(cats_corr$p.value),5)
cats_corr 
```

## Outliers Check
 
We have to check each variable to determinate if it has outliers and which values are considered outlayers

```r
outlierMatrix <- function(data,threshold=1.5) {
    vn <- names(data)
    outdata <- data.frame(row1=1:nrow(data))
    for(v in vn) {
        if(is.numeric(data[[v]])) {
            outlow <- quantile(data[[v]],probs = 0.25,na.rm = T) 
            outhigh <- quantile(data[[v]],probs = 0.75, na.rm = T)
            irq_level <- (outhigh - outlow) * threshold
            outlow <- outlow - irq_level
            outhigh <- outhigh +  irq_level
            mv <- ifelse(data[[v]] < outlow | data[[v]] > outhigh, 1, 0)
            outdata[v] <- mv
        } else {
            mv <- rep(0,nrow(data))
        }
    }
    outdata$row1 <- NULL
    return(outdata)
}
```

#### Checking for differences in the distribution
 
We will compare if there are a difference between the distribution of the variable when having the outliers and when the outliers are removed. We do so using the <b><u>Kolmogorov–Smirnov</u></b> statistic. This is a non-parametric statistic that can be used on variables without outliers.

```r
Rainfall_out <- df$Rainfall
Rainfall_non <- df$Rainfall[which(outmat$Rainfall==0)]
length(Rainfall_out)
length(Rainfall_non)
ks <- ks.test(Rainfall_out, Rainfall_non)
ks
```

We will iterate on each variable with a loop, and will save the results in a data frame. If the difference (p-value) is significant we will add a plus, while when the p-value is not significant, we will add a minus.

```r
res1 <- NULL
for (n in numlst) {
    out <- df[[n]]
    non <- df[[n]][which(outmat[[n]]==0)]
    #outnum <- round((length(out) - length(non))/(length(out) * 100),3)
    outnum <- length(out) - length(non)
    pval <- suppressWarnings(ks.test(out, non)$p.value)
    res1 <- rbind(res1, cbind(var=n, outliers_pct=outnum, distribution_changed=ifelse(pval<0.05,"+","-")))
}
res1
```

The second test we have to make with the outliers is to check if they change the correlation between 
the variable and the outcome variable. A statistical method for checking if two correlation are significantly
different was developed by Diedenhofen and Musch<sup>1</sup>. The method was implemented in an R package called <b><i><a href="https://cran.r-project.org/web/packages/cocor/index.html"> corcor </a></b></i>.

------
<sup>1</sup>  <i> Diedenhofen B, Musch J (2015) cocor: A Comprehensive Solution for the Statistical Comparison of Correlations. PLoS ONE 10(4): e0121945. https://doi.org/10.1371/journal.pone.0121945)</i>

```r
#install.packages("cocor")
library(cocor)
```

```r
Rainfall_out <- df$Rainfall
Rainfall_non <- df$Rainfall[which(outmat$Rainfall==0)]
RainTomorrow_out <- ifelse(df$RainTomorrow=="Yes",1,0)
RainTomorrow_non <- RainTomorrow_out[which(outmat$Rainfall==0)]
df1 <- data.frame(Rainfall_out=Rainfall_out,RainTomorrow_out=RainTomorrow_out)
df2 <- data.frame(Rainfall_non=Rainfall_non,RainTomorrow_non=RainTomorrow_non)

cr <- cocor(~ Rainfall_out + RainTomorrow_out | Rainfall_non + RainTomorrow_non, data=list(df1,df2))
cr
```

```r
cr@fisher1925$p.value
```

```r
res2 <- NULL
for (n in numlst) {
    out <- df[[n]]
    non <- df[[n]][which(outmat[[n]]==0)]
    RainTomorrow_out <- ifelse(df$RainTomorrow=="Yes",1,0)
    RainTomorrow_non <- RainTomorrow_out[which(outmat[[n]]==0)]
    outdf <- data.frame(x_out=out,y_out=RainTomorrow_out)
    nondf <- data.frame(x_non=non,y_non=RainTomorrow_non)
    cr <- cocor(~ x_out + y_out | x_non + y_non, data=list(outdf,nondf))
    pval <- cr@fisher1925$p.value
    res2 <- rbind(res2, cbind(var=n, correlation_changed=ifelse(pval<0.05,"+","-")))
}
res2
```

Now we can join the two dataframes and decide if we can delete the outliers or not!

```r
res <- inner_join(data.frame(res1),data.frame(res2),by="var")
res$drop <- ifelse(res$distribution_changed=="+" & res$correlation_changed == "+","No","Yes")
res
```

# Missing Values

```r
missingMatrix <- function(data) {
    vn <- names(data)
    missdata <- data.frame(row1=1:nrow(data))
    for(v in vn) {
        mv <- ifelse(is.na(data[[v]]),1,0)
        missdata[v] <- mv
    }
    missdata$row1 <- NULL
    return(missdata)
}

getMissingness <- function (data, getRows = FALSE) {
    require(dplyr)
    l <- nrow(data)
    vn <- names(data)
    nadf <- data
    cnt <- NULL
    miss <- function(x) return(sum(is.na(x)))
    for (n in vn) {
        nadf[[n]] <- ifelse(is.na(nadf[[n]]) == T, 1, 0)
        cnt <- rbind(cnt, data.frame(n, sum(nadf[[n]])))
    }
    names(cnt) <- c("var", "na.count")
    cnt$rate <- round((cnt$na.count/nrow(nadf)) * 100, 1)
    nadf$na.cnt <- 0
    nadf$na.cnt <- rowSums(nadf)
    cnt <- cnt %>% dplyr::arrange(desc(na.count)) %>% dplyr::filter(na.count > 
        0)
    totmiss <- nadf %>% dplyr::filter(na.cnt == 0) %>% dplyr::tally()
    idx <- NULL
    msg <- (paste("This dataset has ", as.character(totmiss), 
        " (", as.character(round(totmiss/nrow(data) * 100, 1)), 
        "%)", " complete rows. Original data has ", nrow(data), 
        " rows.", sep = ""))
    if (getRows == TRUE & totmiss != 0) {
        nadf$rn <- seq_len(nrow(data))
        idx <- nadf %>% dplyr::filter(na.cnt == 0) %>% dplyr::select(rn)
    }
    print(list(head(cnt, n = 10), msg))
    return(list(missingness = cnt, message = msg, rows = idx$rn))
}
```

#### Making desicions about the mechanism of missingness using logistic regression
 
Here we run logistic regressions for each of the variables having missing values and check for cases when the p-values for the influence of explaining variables are significant.


```r
### test the missing mechanism using glm
res <- NULL
for (m in mm$missingness$var) {
    ## change the current tested variable with the missing indicator for this variable
    ff <- df[,nm]
    ff[[m]] <- df_na[[m]]
    mod <- glm(ff[[m]] ~.,data=ff,family="binomial")
    sm <- summary(mod)
    if(is.null(sm)==F) {
      sm2 <- data.frame(var=row.names(sm$coefficients),pvalue=sm$coefficients[,4])
      res <- rbind(res, cbind(m,sm2))
      #append(res, sm, m)
    } else {
      print(sm)
    }
}

row.names(res) <- NULL
res
```

## Imputation of the missing values
 
The imputation of the missing values can be done using different techniques. The most frequently used techniques are the substitution of the missing values with a constant (0 or the mean/median of the variable, etc). This technique is not recommended as it can introduce bias into the data.
 
Another technique is known as multiple imputation (MI). This technique uses predictive models to predict the missing values. It begins with the complete variables (those without missing values), and predict the value of the variable with the lower percent of missingness. The model is then ran on the whole dataset and predicted values obtained are used to substitute the original variable. To prevent introduction of bias, it repeats the process many times and generate new datasets. The models have then to be ran on each of this datasets.
 
There are other types of methods like the use of the kNN algorithm. The k-Nearest Neighbours is a model that use the known data to predict the unknown data by positioning the point representing each row in a multidimentional space and then try to check the value of the k-neighbours of the predicted variable and try to determine the approximate value of the missing variable. This technique can be used without the need to make many copies of the data, as is done in MI.
 
Here we will use kNN to make the imputation of the missing values
 
 

```r
# 1. get the name of the complete variable
cv <- setdiff(names(df),mm$missingness$var)
cv <- setdiff(cv,"Date")
cv
```

```r
# 2. get the missing variables ordered by their missingness
```

```r
ord <- mm$missingness %>% arrange(rate)
ord
```

```r
normalizedVars <- function(data) {
    for(v in names(data)) {
        x <- data[[v]]
        if(is.factor(x)) {
            x <- as.numeric(x)
        }
        data[[v]] <- ((x - min(x,na.rm=T)) / (max(x,na.rm=T) - min(x,na.rm=T))) 
    }
    return(data)
}
```

```r
# 3. iterate on variables and fill the missing using kNN
library(class)
df2 <- df

for (i in ord$var) {
    complete <- df2[,c(all_of(cv),i)] %>% filter(is.na(i)==F)
    missdf <-  df2[,c(all_of(cv),i)] %>% filter(is.na(i)==T) %>% select(cv)
    missdf[[i]] <- knn(complete,missdf,cl=i,k=15)
    complete <- rbind(complete,missdf)
    complete <- complete %>% arrange(row.name(complete))
}
```

```r
i="MaxTemp"
df2 <- normalizedVars(df)
complete <- df2[which(is.na(df2[[i]])==F),c(cv,i)]
complete[[i]] <- NULL
class <- df2[which(is.na(df2[[i]])==F),i]
missdf <-  df2[which(is.na(df[[i]])==T),c(cv,i)]# %>% select(cv)
missdf[[i]] <- NULL
```
