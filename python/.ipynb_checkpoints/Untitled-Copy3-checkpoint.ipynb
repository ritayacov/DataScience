##### 2b. How many studenst has the English teacher by course and in total?

```python
df_std_crs = pd.merge(df_Classrooms[['CourseId','StudentId']], df_Courses[['CourseId','CourseName','DepartmentID']], on='CourseId', how='outer')
df_std_crs
```

```python
df_English_f=df_std_crs.query('DepartmentID==1')
df_English_f
```

```python
df_std_English=df_English_f.groupby('CourseName').agg({'StudentId':['count']})
df_std_English
```
