## Q6. For each department, how many students passed with a grades over 80?

```python
df_degree = pd.merge(df_Classrooms[['CourseId','StudentId','degree']], df_Courses[['CourseId','DepartmentID','DepartmentName']], on='CourseId', how='inner')
df_degree80_f=df_degree.query('degree>80')
df_degree80_f
```

```python
df_degree80_f = df_degree80_f.drop_duplicates(['StudentId','DepartmentID','DepartmentName'], keep='last')
df_degree80_f
```

```python
df_degree80_f=df_degree80_f.groupby('DepartmentName').agg({'StudentId':['count']})
df_degree80_f
```

```python
df_80=df_degree80_f.merge(df_std_dep, left_on='DepartmentName', right_on='DepartmentName',
          suffixes=('_80', '_total'))
df_80
```

```python
per=df_degree80_f.div(df_std_dep, level=1, fill_value=0)
per
```

```python
df_per=df_80.merge(per, left_on='DepartmentName', right_on='DepartmentName')
df_per
```

```python
df_per80 = df_per.rename(columns={'StudentId': 'StudentId80_per'})
df_per80
```

## Q10. Create a dataframe showing the students, the number of courses they take, the average of the grades per class, and their overall average (for each student show the student name).

```python
CRS_num=df_Classrooms.groupby('StudentId').agg(['count'])['CourseId']                                                   
CRS_num
```

```python
CRS_num1=pd.DataFrame({'StudentId': CRS_num.index.to_list(),
                                'CourseId': CRS_num['count'].to_list()})
CRS_num1
```

```python
General_Degree=df_Classrooms.groupby('StudentId').agg(['mean'])['degree']                                                   
General_Degree
```

```python
General_Degree1=pd.DataFrame({'StudentId': General_Degree.index.to_list(),
                               'degree': General_Degree['mean'].to_list()})
General_Degree1
```

```python
df_Deg_STD=pd.merge(CRS_num1[['StudentId','CourseId']], General_Degree1[['StudentId','degree']], on='StudentId', how='outer')
df_Deg_STD
```

```python
df_crs_std1 = pd.merge(df_Classrooms[['StudentId','CourseId','degree']], df_Courses[['CourseId','DepartmentName']], on='CourseId', how='outer')
df_crs_std1.drop_duplicates()
```

```python
df_crs_std2= pd.merge(df_crs_std1[['StudentId','CourseId','degree','DepartmentName']], df_Students[['StudentId','FirstName','LastName']], on='StudentId', how='outer')
df_crs_std2.drop_duplicates()
```

```python
df_crs_std3 = pd.merge(df_crs_std2[['StudentId','FirstName','LastName','degree','DepartmentName']], df_Deg_STD[['StudentId','CourseId','degree']], on='StudentId', how='outer')
df_crs_std3
```

```python
df_crs_std4 = df_crs_std3.rename(columns={'CourseId': 'Course_num', 'degree_y': 'General_Degree', 'degree_x': 'Mean_degree'})
df_crs_std4
```

```python
df_crs_std5=df_crs_std4.groupby(['StudentId','FirstName','LastName','DepartmentName','Course_num','General_Degree']).mean()
df_crs_std5.dropna()
```

```python
df_crs_std6=pd.pivot_table(df_crs_std5, values=['Mean_degree'], index=['StudentId','FirstName','LastName','Course_num','General_Degree'], columns=['DepartmentName'])
df_crs_std6
```
