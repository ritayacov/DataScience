# Questions
 
 
## Q1. Count the number of students on each department

```python
stud_class = pd.merge(students,classrooms,how='inner',on='StudentId')
stud_class.head()
```

```python
stud_cour = pd.merge(stud_class,courses,how='inner', on='CourseId')
stud_cour.head()
```

```python
stud_depr = pd.merge(stud_cour,departments,how='inner', on='DepartmentID')
stud_depr.head()
```

```python
q1 = stud_depr.groupby('DepartmentName').agg({'StudentId':['count']})
q1
```

## Q2. How many students have each course of the English department and the total number of students in the department?

```python
q2 = stud_depr[stud_depr.DepartmentName == 'English']
q2
```

```python
q2a = q2.groupby('CourseName').agg({'StudentId':['count']})
q2a
```

```python
q2b = q2.groupby('DepartmentID').agg({'StudentId':['count']})
q2b
```

## Q3. How many small (<22 students) and large (22+ students) classrooms are needed for the Science department?

```python
q3 = stud_depr[stud_depr.DepartmentName == 'Science']
q3
```

```python
q3a = q3.groupby('CourseName').agg(['count'])['StudentId']
q3a
```

```python
q3a.dtypes
```

```python
q3a.columns
```

```python
q3b=pd.DataFrame({'CourseName':q3a.index.to_list(),
                 'StudentId' : q3a['count'].to_list()})
q3b
```

```python
q3b.dtypes
```

```python
q3b.columns
```

```python
q3b['ClassSize'] = 'small'
q3b['ClassSize'].head()
q3b.loc[(q3b['StudentId']<=22),'ClassSize']='small'
q3b.loc[(q3b['StudentId']>22),'ClassSize']='big'
q3b
```

```python
q3b.groupby('ClassSize').agg({'StudentId':['count']})
```
